```python
days=[1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.legend(loc='best')
```

```python
import numpy as np
```

```python
company =['Google', 'Amazon', 'Microsft', 'Meta']
revenue = [90,136,89,59]
profit=[40,2,32,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue')
plt.title('US tech stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue')
plt.title('US tech stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.bar(ypos, profit, label='Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue')
plt.title('US tech stocks')
plt.bar(xpos-0.2, revenue,width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit,width=0.4, label='Profit')
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.ylabel('revenue')
plt.title('US tech stocks')
plt.barh(ypos-0.2, revenue, height=0.4,label='Revenue')
plt.barh(ypos+0.2, profit, height=0.4, label='Profit')
plt.legend()
```

```python
blood_sugar = [112,85,150,139,88,92,114,80,77,82,129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [112,85,150,139,88,92,114,80,77,82,129]
plt.hist(blood_sugar, bins=3,rwidth=0.95)
```

```python
blood_sugar_men = [112,85,150,139,88,42,114,80,77,82,129]
blood_sugar_women = [114,84,130,109,48,32,134,30,67,52,139]
plt.hist([blood_sugar_men, blood_sugar_women], bins=3,rwidth=0.95,label=['men','women'])
plt.xlabel('sugar range')
plt.ylabel('total no of patients')
plt.title('blood sugar')
plt.legend()
```

```python
blood_sugar_men = [112,85,150,139,88,42,114,80,77,82,129]
blood_sugar_women = [114,84,130,109,48,32,134,30,67,52,139]
plt.hist([blood_sugar_men, blood_sugar_women], bins=3,rwidth=0.95,label=['men','women'], orientation='horizontal')
plt.xlabel('sugar range')
plt.ylabel('total no of patients')
plt.title('blood sugar')
plt.legend()
```

```python
exp_vals = [1400,500,300,410,259]
exp_labels = ['home rent','food','phone','car','bills']
```
